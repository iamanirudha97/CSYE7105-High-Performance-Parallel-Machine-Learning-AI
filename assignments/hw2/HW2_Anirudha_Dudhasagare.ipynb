# **Part 2**

```python
# Part_2
# Q.1
def part2(df):
  return (df - df.mean())/df.std()
  
```

```python
# Q2. Parallelize this function “part2”:

from multiprocessing import Pool, cpu_count

def part2_parallel(df, cores):
  chunks = np.array_split(df, cores*2, axis=1)

  start = time.time()
  with Pool(processes=cores) as pool:
    normalized_chunks = pool.map(part2, chunks)
  result = pd.concat(normalized_chunks, axis=1)
  end = time.time()
  duration = end - start
  print(f"{cores} CPUs ----> {duration:.4f} seconds")
  return result, duration
```

```python
# Q.2.2
cpu_cores = [1, 2, 4, 6, 8]
time_elapsed = []
for core in cpu_cores:
  _, t = part2_parallel(df, core)
  time_elapsed.append(t)
  print("--------------------------------------------------------------------------------------")

time_elapsed
```

```python
# Calculating speedup and efficiency for all cores
speed_ups = [time_elapsed[0] / t for t in time_elapsed]
efficiency = [speed_ups[i]/cpu_cores[i] for i in range(0, len(speed_ups))]

efficiency, speed_ups
```

```python
# Q.2.3 and Q.2.4
import matplotlib.pyplot as plt

def plot_graphs(cpu_cores, time_elapsed, speed_ups, efficiency, tag):
  # ------------------ Plot 1: Wall-Clock Time ------------------
  plt.figure(figsize=(6,4))
  plt.plot(cpu_cores[1:], time_elapsed[1:], marker='o', color='orange')
  plt.title('Wall-Clock Time vs CPU Count')
  plt.xlabel('Number of CPUs')
  plt.ylabel('Time (seconds)')
  plt.grid(True)
  plt.tight_layout()
  plt.savefig(f'{tag}_wall_clock_time.png', dpi=300)
  plt.show()

  # ------------------ Plot 2: Speedup ------------------
  plt.figure(figsize=(6,4))
  plt.plot(cpu_cores[1:], speed_ups[1:], marker='o', color='steelblue')
  plt.title('Speedup vs CPU Count')
  plt.xlabel('Number of CPUs')
  plt.ylabel('Speedup (T₁ / Tₙ)')
  plt.grid(True)
  plt.tight_layout()
  plt.savefig(f'{tag}_speedup.png', dpi=300)
  plt.show()

  # ------------------ Plot 3: Efficiency ------------------
  plt.figure(figsize=(6,4))
  plt.plot(cpu_cores[1:], efficiency[1:], marker='o', color='green')
  plt.title('Efficiency vs CPU Count')
  plt.xlabel('Number of CPUs')
  plt.ylabel('Efficiency (Speedup / N)')
  plt.grid(True)
  plt.tight_layout()
  plt.savefig(f'{tag}_efficiency.png', dpi=300)
  plt.show()

  print("saved all three pngs'")

plot_graphs(cpu_cores, time_elapsed, speed_ups, efficiency, "Part2_Q2")
```

```python
# Q.3
def part2_parallel_imap(df, cores, chunk_size):
  chunks = np.array_split(df, cores*2, axis=1)
  start = time.time()

  with Pool(processes=cores) as pool:
    res = pool.imap(part2, chunks, chunksize=chunk_size)
    normalized_chunks = list(res)

  result = pd.concat(normalized_chunks, axis=1)
  end = time.time()
  duration = end - start
  print(f"{cores} CPUs ----> {duration:.3f} seconds")
  return result, duration
```

```python
# Q.3.2
chunk_sizes = [3, 7]
imap_results = {}
for chunk in chunk_sizes:
    times = []
    print(f"for chunck size: {chunk}")
    for c in cpu_cores:
        _, duration = part2_parallel_imap(df, c, chunk)
        times.append(duration)
    imap_results[chunk] = times

imap_results
```

```python
# Q.3.3
imap_speedup = {}
imap_efficiency = {}

for chunk in chunk_sizes:
  base_time = imap_results[chunk][0]
  speedup = [base_time / t for t in imap_results[chunk]]
  efficiency = [speedup[i] / cpu_cores[i] for i in range(len(cpu_cores))]

  imap_speedup[chunk] = speedup
  imap_efficiency[chunk] = efficiency

imap_speedup, imap_efficiency

# for chunk in imap_results.keys():
#     times = imap_results[chunk]
#     speedup = imap_speedup[chunk]
#     efficiency = imap_efficiency[chunk]

#     print(f"chunk {chunk}, times: {times}, speedup: {speedup}, efficiency: {efficiency}")
```

```python
# Q.3.4
import matplotlib.pyplot as plt

def plot_imap_graphs(cpu_cores, imap_results, imap_speedup_results, imap_efficiency_results, tag):

    # ---- Wall-clock Time ----
    plt.figure()
    for chunk, times in imap_results.items():
        plt.plot(cpu_cores[1:], times[1:], marker='o', label=f'chunksize={chunk}')
    plt.title('Wall-clock Time vs CPUs')
    plt.xlabel('Number of CPUs')
    plt.ylabel('Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{tag}_imap_time.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ---- Speedup ----
    plt.figure()
    for chunk, speedup in imap_speedup_results.items():
        plt.plot(cpu_cores[1:], speedup[1:], marker='o', label=f'chunksize={chunk}')
    plt.title('Speedup vs CPUs')
    plt.xlabel('Number of CPUs')
    plt.ylabel('Speedup (T₁ / Tₚ)')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{tag}_imap_speedup.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ---- Efficiency ----
    plt.figure()
    for chunk, efficiency in imap_efficiency_results.items():
        plt.plot(cpu_cores[1:], efficiency[1:], marker='o', label=f'chunksize={chunk}')
    plt.title('Efficiency vs CPUs')
    plt.xlabel('Number of CPUs')
    plt.ylabel('Efficiency (Speedup / CPUs)')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{tag}_imap_efficiency.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("Saved combined comparison plots for all chunk sizes.")


plot_imap_graphs(cpu_cores, imap_results, imap_speedup, imap_efficiency, "Part2_Q.3")
```
